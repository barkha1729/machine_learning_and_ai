## 2.1 With Data Dictionary

 Adding the data dictionary with loan data based on the column names

```python
def study_cols_with_descriptions(data_dict_df, loan_df):
    ''' This function combines loan data with its data dictionary
    params:
        data_dict_df: df containg the first sheet of data dictionary excel
        loan_df: df containing the data from loan.csv

    returns:
        data_with_def_df: df with column_names,dtypes,sample_data,column description        
    '''
    
    print("Joining loan column's data and dtype with column descriptions \n")
    
    # loan_df_dtypes: extracting data types of the loan_df in this new df loan_df_dtypes 
    loan_df_dtypes = pd.DataFrame(loan_df.dtypes,columns=['type'])
    loan_df_dtypes = loan_df_dtypes.reset_index()
    loan_df_dtypes['col_name'] = loan_df_dtypes['index']
    loan_df_dtypes = loan_df_dtypes[['col_name','type']]

    # adding a sample value from data (first row's value) to loan_df_dtypes
    loan_df_dtypes['sample_data'] = loan_df.loc[0].values 

    print(str(len(loan_df.columns))+" columns present in loan.csv")
    print(str(len(data_dict_df))+" columns present in data dictionary")
    
    #  joining loan_df_dtypes with data dictionary
    data_with_def_df = pd.merge(loan_df_dtypes,data_dict_df,how='left',
                                left_on='col_name',right_on='LoanStatNew' )
    data_with_def_df.set_index(["col_name"], inplace = True, drop = True)
    print("\n Added data dictionary to data!")
    return data_with_def_df.T
```

```python
# if data dictionary exists joining with loan.csv
data_with_def_df = None
if os.path.exists(data_dictionary_file_path):
    data_dict_df = pd.read_excel(data_dictionary_file_path,sheet_name=0)
    loan_df_raw_copy = loan_df_raw.copy()
    data_with_def_df = study_cols_with_descriptions(data_dict_df, loan_df_raw_copy)
```

```python
def describe_column_definition(loan_df, col_name_list, desc = True, check_unique=False, ignore_nan=True,see=False):
    '''Customised function for describing and analysing based on rules '''
    cols_to_remove = []
    cols_to_decide = []
    
    for col_name in col_name_list:
        col_details = data_with_def_df[col_name]
    
        print("\nName: " + col_details["LoanStatNew"])
        print("Type: " + str(col_details["type"]))
        print("Sample Data: " + str(col_details["sample_data"]))
        print("Description: " + col_details["Description"] +"\n")
        
       
    
        if check_unique:
            
            '''
            Rule: column to be checked should not have all unique values, or one unique value, or too many unique values 
            '''
            
            unique_values = loan_df[col_name].unique()
            if ignore_nan:
                unique_values = [x for x in unique_values if str(x)!="nan" ]
            else:
                no_of_missing_values = loan_df[col_name].isnull().sum()
                perc_missing_values = round(100*loan_df[col_name].isnull().sum()/len(loan_df),2)
                print("Total Missing values: "+ str(no_of_missing_values))
                print("Percentage of Missing values: "+ str(perc_missing_values)+" %")
            
            no_of_unique = len(unique_values)
            total_rows = len(loan_df)
        
            if no_of_unique == total_rows:
                cols_to_remove.append(col_name)
                print("\nNote: Each row has a unique value in "+col_name+". NOT useful for our analysis. \n")
            
            elif no_of_unique == 1:
                cols_to_remove.append(col_name)
                print("\nNote:Only 1 unique value ("+str(unique_values)+") present  in "+col_name+", NOT useful for our target analysis. \n")
            
            elif no_of_unique > 50:
                cols_to_decide.append(col_name)
                print("\nTotal unique values: " +str(no_of_unique))
                print("\nNote: More than 50 unique values present in "+col_name+", DECIDE based on business usecase.\n      Check if this column can be grouped or binned \n")
            
            else:
                print("Unique Values: ")
                print(unique_values)
                print("\nNote: Column looks good for analysis. GO for further analysis. \n")
        
        if see:
            # to see first 5 unique values
            unique_values = loan_df[col_name].unique()
            unique_values = [x for x in unique_values if str(x)!="nan" ]
            print(unique_values[:5])
    
    if check_unique: 
        if len(cols_to_remove) != 0 and len(cols_to_decide) != 0 : 
            return cols_to_remove, cols_to_decide
            
        
    return 
```

```python
def univariate_analysis(col_name, full_name, continuous, sort = True, orientation= 'horizontal'):
    """
    Customised function for visualizing a column using loan_df
    - col_name is the colum name in the dataframe to be analysed
    - full_name is the full name of the column
    - continuous is True if the variable is continuous, False otherwise
    - sort : if true then sort the bars
    """
    if orientation == 'vertical':
        rows = 2
        cols = 1
        f, ((ax1), (ax2)) = plt.subplots(nrows=rows, ncols=cols, figsize=(12,5), dpi=90)
        plt.xticks(rotation = 90)
        
    else:
        rows =1
        cols =2
        f, (ax1, ax2) = plt.subplots(nrows=rows, ncols=cols, figsize=(12,3), dpi=90)
    
    #plt.xticks(rotation=orientation)
    # Plot without loan status
    if continuous:
        sns.distplot(loan_df.loc[loan_df[col_name].notnull(), col_name], kde=False, ax=ax1)
    else:
        if sort:
            sns.countplot(loan_df[col_name], order=sorted(loan_df[col_name].unique()), color='#5975A4', saturation=1, ax=ax1)
        else:
            sns.countplot(loan_df[col_name], order=loan_df[col_name].unique(), color='#5975A4', saturation=1, ax=ax1)
    ax1.set_xlabel(full_name)
    ax1.set_ylabel('Count')
    ax1.set_title(full_name)
    
    # Plot with loan status
    if continuous:
        sns.boxplot(x=col_name, y='is_defaulter', data=loan_df, ax=ax2)
        ax2.set_ylabel('')
        ax2.set_title(full_name + ' by Loan Status')
    else:
        charge_off_rates = loan_df.groupby(col_name)['is_defaulter'].value_counts(normalize=True).loc[:,'Defaulter']
        sns.barplot(x=charge_off_rates.index, y=charge_off_rates.values, color='#5975A4', saturation=1, ax=ax2)
        ax2.set_ylabel('Fraction of Defaulters')
        ax2.set_title('Default Rate by ' + full_name)
    ax2.set_xlabel(full_name)
    
    plt.tight_layout()
```

```python
def get_year(value):
    if isinstance(value,str):
        new_value = value.split("-")[1]
        if int(new_value) > 40:
            return int(new_value)+ 1900
        else:
            return int(new_value)+ 2000
    else:
        return value
```

```python
def bivariate_analysis(col1,col2, title, bar=False):

    two_var_df = bivariate_loan_df.loc[:,[col1,col2,'defaulter_count']]
    
   
    two_var_df = two_var_df.groupby(by=[col1,col2]).agg({'defaulter_count':['sum','count']})
    two_var_df["perc_defaulter"] =round(100*two_var_df[("defaulter_count","sum")]/two_var_df[("defaulter_count","count")],2)
    two_var_df.reset_index(inplace=True)
    two_var_df =two_var_df.sort_values(by=['perc_defaulter'],ascending = False)
    two_var_df =two_var_df.sort_values(by=[col1]) 
    
    if bar:
        two_var_plot = sns.barplot(x=col1, y='perc_defaulter', hue=col2, data=two_var_df)
    else:
        two_var_plot = sns.lineplot(x=col1, y='perc_defaulter', hue=col2, data=two_var_df)
    plt.title(title)
    # Put the legend out of the figure
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.xticks(rotation=45)
    two_var_df_pivoted = two_var_df.pivot(index =col1, columns =col2, values =['perc_defaulter'])
    print(two_var_df_pivoted)
    return 
```

```python

title='Percentage of defaulters in each years of employment by house  ownership'
bivariate_analysis("emp_length","home_ownership",title)
```

```python
# Observation:
# In 1-4 years employed, the percentages of defaulters is maximum among the OTHER category in home ownership
# profiles belonging to MORTGAGE in house ownership are less likely to default than any other category excluding OTHER
# Percentage of defaulters is very less for people with 4,6,8 years of employment years and belonging to OTHER category
```
